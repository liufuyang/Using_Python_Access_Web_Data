```python
import urllib
import xml.etree.ElementTree as ET
```

```python
serviceurl = 'http://python-data.dr-chuck.net/comments_288665.xml'
```

```python
uh = urllib.urlopen(serviceurl)
data = uh.read()
```

```python
tree = ET.fromstring(data)
```

```python
sum(map(lambda e: int(e.text), tree.findall('.//count')))
```

```python
import json
import urllib

serviceurl = 'http://python-data.dr-chuck.net/comments_288669.json'

uh = urllib.urlopen(serviceurl)
data = uh.read()
info = json.loads(data)
sum(map(lambda c: c['count'], info['comments']))
```
